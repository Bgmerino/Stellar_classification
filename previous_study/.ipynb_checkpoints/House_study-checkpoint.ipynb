## Use Vector Autoregression (VAR)

VAR models are the best option when dealing with multivariate models. They operate with only one operation per time step, so we will need to group the data to get only one variable value per time point. We will do it by mean


```python
df_group = df_ini.groupby('datesold')[['price', 'bedrooms']].mean().reset_index()

df_group
```

```python
plt.plot(df_group['datesold'], df_group['price'])

plt.xlabel('Date')
plt.ylabel('Price')
xticks = df_group['datesold'][::200]

plt.xticks(xticks, rotation=45)

plt.show()
```

```python
plt.plot(df_group['datesold'], df_group['bedrooms'])

plt.xlabel('Date')
plt.ylabel('Bedrooms')
xticks = df_group['datesold'][::200]

plt.xticks(xticks, rotation=45)
plt.show()
```

The data will be splited in two parts: training and testing. We will use the majority of the data for training leaving the last month to test the model

```python
n_obs = 30 
df_train, df_test = df_group[:-n_obs], df_group[-n_obs:]

len(df_train),len(df_test)
```

```python
plt.plot(df_train['datesold'], df_train['bedrooms'])
plt.plot(df_test['datesold'], df_test['bedrooms'])


plt.xlabel('Date')
plt.ylabel('Bedrooms')

plt.show()
```

```python
plt.plot(df_train['datesold'], df_train['price'])
plt.plot(df_test['datesold'], df_test['price'])

plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

### Stationary check

To make sure that the model would be able to predict the future steps, the time series needs to present constant properties over time, be stationary. To test so Augmented Dickey-Fuller test is used.
If p- value < 0.5 the series is likely stationary.

```python
def check_stationarity(df, name=''):
    result = adfuller(df.dropna())
    print('ADF Test for {}'.format(name))
    print('  ADF Statistic : {}'.format(result[0]))
    print('  p-value       : {}'.format(result[1]))
```

```python
check_stationarity(df_train['price'], name='price')
check_stationarity(df_train['bedrooms'], name='bedrooms')
```

This time series is clearly stacionary as p value is really small, almost cero and ADF Statistic is minor that the value to the 1%
 1%: -3.50
 5%: -2.89
10%: -2.58

So it won't be necessary to differenciate the time series.

```python
df_train = df_train.set_index('datesold')  
df_test = df_test.set_index('datesold')  
```

### Lag order

It is necessary to select the optimal lag order to train the model. To do so we will fit the model considering different lags and then they will evaluated considering two criteria: AIC and BIC 

```python
model = VAR(df_train)
for i in [1,5,10,15,100,150]:
    result = model.fit(i)
    print('Lag Order =', i)
    print('AIC : ', result.aic)
    print('BIC : ', result.bic)
```

Looking at this values, we can see how until lag 15 the model improves both metrics, but from there on the value of BIC stars increasing. This is because as the lag increases the nodel fits the training better, getting a better value of AIC. However the odel becomes more complicates penalizing BIC.
To obtain a good accuracy and to avoid underfitting, we are going to select the lag that gives the best AIC value: 100 

```python
model = VAR(df_train)
model_f = model.fit(100)

model_f.summary()
```

### Serial Correlation Residuals

We are going to look for autocorrelation in the residuals of the model, to check if the model has fully explained the time-dependent structure in the data. 

To do so Durbin Watson’s Statistic is used. Its value can vary between 0 and 4, being 2 the value for no significant serial correlation. The closer to 0, there is a positive serial correlation, and the closer it is to 4 implies negative serial correlation.

Having a value of 2 shows that the model is no leaving any important structure withoiut conideration

```python
out = durbin_watson(model_f.resid)

for col, val in zip(df_group.columns[1:], out):
    print((col), ':', round(val, 2))
```

### Forecast

Now that the model is trained we are going to use it to predict over the test data, where we will evaluate its performance. We will consider a 30 days period 

```python
lag_order = model_f.k_ar
nobs = (df_test.index[-1]-df_test.index[0]).days+2
nobs
```

```python
forecast_input = df_train.values[-lag_order:]
```

```python
forecast = model_f.forecast(y=forecast_input, steps=nobs)
```

```python
df_forecast = pd.DataFrame(forecast, 
                           index=pd.date_range(start=df_train.index[-1], periods=nobs+1, freq='D')[1:],
                           columns=df_test.columns)
```

### Forecast evaluation

To evaluate the prediction, the error between the test data and the forecst is compared, obtaining the Mean absolute Error (mae) and the Root Mean Squared Error (rmse).

```python
filtered_df = df_forecast[df_forecast.index.isin(df_test.index)]
```

```python
mae = mean_absolute_error(df_test[['price']], filtered_df[['price']])
rmse = np.sqrt(mean_squared_error(df_test[['price']], filtered_df[['price']]))
print('MAE: {}, RMSE: {}'.format(mae,rmse))
```

Even though the model predicts on all the variables provided, bedrooms and price, we will consider only the price as it iis the main variable of the study

The modelling is following the evolution of the data, but it not really catching its variations.

```python
plt.plot(df_test.index,df_test['price'],label='test')
plt.plot(df_forecast.index,df_forecast['price'],label='prediction')

xticks = df_forecast.index[::2]
plt.xticks(xticks, rotation=45)

plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

```

As the data was stationary differentiation wasn't needed, so now that step doesn't need to be undone. 

## ARIMA

With the aim of improving the prediction, ARIMA model will be tested. SARIMA is not being considered as the selected data has no seasonality.

```python
df_group = df_ini.groupby('datesold')[['price', 'bedrooms']].mean().reset_index()
```

```python
n_obs = 30 
df_train_, df_test_ = df_group[:-n_obs], df_group[-n_obs:]
```

```python
df_train_ind=df_train_.set_index('datesold')
df_train_ind.index = pd.to_datetime(df_train_ind.index)

df_test_ind=df_test_.set_index('datesold')
df_test_ind.index = pd.to_datetime(df_test_ind.index)

df_train_ind
```

### Determining ARIMA parameters

To fit an ARIMA model two parameters need to be stablished: 
- the moving average (q) to determine how many past forecast errors need to be included
- the autoregressive (p) to know how many past values need to be included
- differentiation (d) is the number of times the data needed to be differentiated to obtain a estationary behaviour. 

As the data was stationary, d=0.
To determine q the plot ACF will be considered and for p the PACF plot 

#### ACF Plot

```python
plt.figure(figsize=(12, 6))

plot_acf(df_train_ind['price'], lags=50)  # Adjust lags as needed
plt.title('Autocorrelation of House Prices')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```

Looking at the graph, the best value is obtain for 6, so this will be the value used for q.

#### PACF Plot

```python
plot_pacf(df_train_ind['price'], lags=50)  # Adjust lags as needed
plt.title('Autocorrelation of House Prices')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```

In this case, the best autocorrelation is also found for p=6

```python
model = ARIMA(df_train_ind['price'], order=(6,0,6))  
model_fit = model.fit()
print(model_fit.summary())
```

### Serial Correlation Residuals

```python
out = durbin_watson(model_fit.resid)
print('price: {}'.format(out, 2))
```

### Forecast

As we did for the model VAR we will predict to compare with the test data we have

```python
forecast_steps = (df_test_ind.index[-1]-df_test_ind.index[0]).days+2
forecast = model_fit.get_forecast(steps=forecast_steps)
```

```python
forecast_steps
```

```python
forecast_values = forecast.predicted_mean
forecast_conf_int = forecast.conf_int()
```

```python
last_date = df_train_ind.index[-1]
last_date
```

```python
last_date = df_train_ind.index[-1]
forecast_index = pd.date_range(last_date, periods=forecast_steps+1, freq="D")[1:]
forecast_index
```

Once the forcast is calculated we plot the data with the test values for comparisson

```python
plt.plot(df_test_ind.index,df_test_ind['price'], label='Original Data', color='blue')
plt.plot(forecast_index, forecast_values, label='Forecast', color='orange')
plt.fill_between(
    forecast_index,
    forecast_conf_int.iloc[:, 0],
    forecast_conf_int.iloc[:, 1],
    color="orange",
    alpha=0.3,
    label='Confidence Interval'
)
plt.title('ARIMA Model Forecast', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(xticks, rotation=45)
plt.tight_layout()
plt.show()
```

### Forecast evaluation

```python
filtered_df = forecast_values[forecast_index.isin(df_test_ind.index)]
```

```python
mae = mean_absolute_error(df_test_ind[['price']], filtered_df)
rmse = np.sqrt(mean_squared_error(df_test_ind[['price']], filtered_df))
print('MAE: {}, RMSE: {}'.format(mae,rmse))
```

The errors obtained here, are slightly bigger than the ones obtained for the VAR model. 
MAE: 55792.58, RMSE: 176521.06

## ARIMA on a new variable

We have tried to predict the general price of the houses, but we haven't paid so much attention to their size. To take also this variable into account, we are going to predict the house price per room

It is not the goal of the exercise, but the data could be fitted with a linear trend to under 

```python
df_ini['price_room']=df_ini['price']/df_ini['bedrooms']
df_ini
```

We need to correct the cases in which the house doesn't have any rooms, in that case we use the price of teh house

```python
df_ini['price_room'].replace([np.inf, -np.inf], np.nan, inplace=True)

df_ini['price_room'] = df_ini['price_room'].fillna(df_ini['price'])
```

We group the data, divide it in train and test, and check that it is still stationary

```python
df_group2 = df_ini.groupby('datesold')[['price_room']].mean().reset_index()
df_group2
```

```python
n_obs=30
df_train2, df_test2 = df_group2[:-n_obs], df_group2[-n_obs:]
```

### Stationary check

As the data has change, because a new variable has been calculated, let's check that it is stationary

```python
check_stationarity(df_train2['price_room'], name='price_room')
```

```python
df_train_ind2=df_train2.set_index('datesold')
df_train_ind2.index = pd.to_datetime(df_train_ind2.index)

df_test_ind2=df_test2.set_index('datesold')
df_test_ind2.index = pd.to_datetime(df_test_ind2.index)

df_train_ind2
```

### Determining ARIMA parameters

#### ACF Plot

```python
plot_acf(df_train_ind2['price_room'], lags=30)  # Adjust lags as needed
plt.title('Autocorrelation of House Prices')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```

#### PACF Plot

```python
plot_pacf(df_train_ind2['price_room'], lags=30)  # Adjust lags as needed
plt.title('Autocorrelation of House Prices')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```

The parameters for the model hasn't change: q=6, d=0, p=6

### Serial Correlation Residuals

This value is slightly smaller than 2, but still pretty close so I will consider that the model is not leaving some structure out
